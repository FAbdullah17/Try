```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_folder_path = "cat_dog"
sample_images = os.listdir(image_folder_path)[:10]
```

```python
plt.figure(figsize=(10, 10))
for idx, image_name in enumerate(sample_images):
    img = Image.open(os.path.join(image_folder_path, image_name))
    plt.subplot(1, 10, idx + 1)
    plt.imshow(img)
    plt.title(image_name)
    plt.axis("off")

plt.show()
```
